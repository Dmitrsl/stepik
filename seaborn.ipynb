```python
event = pd.read_csv('event_data_train.csv')
event.head()
```

```python
sub = pd.read_csv('submissions_data_train.csv')
sub.head()
```

```python
sub['date'] = pd.to_datetime(sub.timestamp, unit='s')
sub['day'] = sub.date.dt.date
event['date'] = pd.to_datetime(event.timestamp, unit='s')
event['day'] = event.date.dt.date
```

```python
event
```

```python
event[['user_id', 'day', 'timestamp']].drop_duplicates(['user_id', 'day']).head()
```

```python
event_max = event.groupby('user_id', as_index=False) \
    .agg({'timestamp': 'max'})
event_max.head()
```

```python
event_min = event.groupby('user_id', as_index=False) \
    .agg({'timestamp': 'min'})
event_min.head()
```

```python
event_max['min'] = event_min['timestamp']
event_max['diff'] = event_max['timestamp'] - event_max['min']
event_max.sort_values(['diff', 'min'], ascending=[False, True]).head(100)
```
