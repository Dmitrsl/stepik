```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
cats_dogs = pd.read_csv('dogs_n_cats.csv')
cats_dogs.head()
```

```python
cats_dogs = pd.get_dummies(cats_dogs)
cats_dogs
```

```python
cats_dogs['Y'] = 0
cats_dogs['Y'][cats_dogs['Вид'] == 'собачка'] = 1
cats_dogs.head()
```

```python
X = cats_dogs.drop(['Y', 'Вид'], 1)
X.head()
```

```python
Y = cats_dogs['Y']
Y.head()
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf.fit(X, Y)
```

```python
clf.score(X_train, y_train), clf.score(X_test, y_test)
```
